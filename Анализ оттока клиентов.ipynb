### Дубликаты

```python
# проверим наличие полных дубликатов и сразу долю отточных в них (если есть)
print('Доля полных дубликатов: {}'.format(round(data['userid'].loc[data.duplicated()].count()/data['userid'].count(), 3)))
print('Доля отточных клиентов в полных дубликатах: {}'.format(data['churn'].loc[data.duplicated()].mean()))
print('')
#проверим, есть ли дубликаты среди уникальных айди клиентов
print('Доля дубликатов user id: {}'.format(round(data['userid'].loc[data['userid'].duplicated()].count()/data['userid'].count(), 3)))
print('Доля отточных клиентов в дубликатах по user id: {}'.format(data['churn'].loc[data['userid'].duplicated()].mean()))
```

Дубликаты среди уникальных айди клиентов - явно техническая ошибка, такие строки стоит удалить, тем более что они занимают меньше 1% от всего датафрейма и не имеют содержат отточных клиентов.

```python
# удалим дубликаты
data = data.drop_duplicates(subset = 'userid')
```

### Пропуски

```python
# выведем долю пропусков
data.isna().mean()
```

Больше 20% пропущенных значений в данных о балансе на счету, меньше 1% пропущенных значений в данных о возрасте клиента.

Пропуски в столбце с возрастом занимают очень маленькую долю от общего количества значений.

Пропуски в столбце с балансом могут быть обозначением нулевого баланса.

#### Пропуски значений возраста клиентов

```python
#выделим количество клиентов с пропущенным возрастом и долю отточных среди них
print('Количество пользователей с пропущенным возрастом: {}'.format(data['userid'].loc[data['age'].isna()].count()))
print('Доля отточных клиентов среди юзеров с пропущенным возрастов: {}'.format(data['churn'].loc[data['age'].isna()].mean()))
```

Таким количеством можно пожертвовать, так как данные о возрасте понадобятся нам для дальнейшего анализа оттока клиентов по различным признакам.

```python
#удалим пропуски в данных о возрасте
data = data.query('~age.isna()')

#изменим тип данных столбца age на целые числа
data['age'] = data['age'].astype(int)
```

### Категоризация данных

На основе анализа распределения категоризуем данные о возрасте, заработной плате и балансе.

#### Возраст

Ранее по графику распределения мы выделили более отточных клиенто в возрасте от ~25 до ~40, от ~50 до ~60. Таким образом можно выделить следующие категории:
1. 18-25
2. 26-40 - самая отточная категория
3. 41-50 
4. 51-60 - также отточная категория
5. 61+

```python
# напишем функцию, которая категоризирует возраст
def age_cat (x):
    if x <= 25:
        return '18-25'
    elif x > 25 and x <= 40:
        return '26-40'
    elif x > 40 and x <= 50:
        return '41-50'
    elif x > 50 and x <= 60:
        return '51-60'
    else:
        return '61+'
    
data['age_cat'] = data['age'].apply(age_cat)
data.head()
```

#### Баланс

Для категоризации баланса также на основе графика распределения выделим следующие категории:
1. <250k
2. 250k-500k
3. 500k-700k
4. 700k-1m - самая отточная группа
5. 1m-1.5m - также отточная группа
6. 1.5m+

```python
# напишем функцию, которая категоризирует баланс
def balance_cat (x):
    if x < 250000:
        return '<250k'
    elif x >= 250000 and x < 500000:
        return '250k-500k'
    elif x >= 500000 and x < 700000:
        return '500k-700k'
    elif x >= 700000 and x < 1000000:
        return '700k-1m'
    elif x >= 1000000 and x < 1500000:
        return '1m-1.5m'
    else:
        return '1.5m+'

data['balance_cat'] = data['balance'].apply(balance_cat)
data.head()
```

#### Зарплата

На основе распределения выделим следующие категории:
1. <90k
2. 90k-150k - самая отточная группа
3. 150k-225k - также отточная группа
4. 225k-350k
4. 350k+

```python
# напишем функцию, которая категоризирует зарплату
def salary_cat (x):
    if x < 90000:
        return '<90k'
    elif x >= 90000 and x < 150000:
        return '90k-150k'
    elif x >= 150000 and x < 225000:
        return '150k-225k'
    elif x >= 225000 and x < 350000:
        return '225k-350k'
    else:
        return '>350k'

data['salary_cat'] = data['est_salary'].apply(salary_cat)
data.head()
```

## Статистический анализ данных

В ходе статистического анализа данных мы проверим две гипотезы:
1. Гипотезу различия дохода между теми клиентами, которые ушли и теми, которые остались
2. Гипотезу различия баланса между теми клиентами, которые ушли и теми, которые остались

Для проверки гипотез воспользуемся t-критерием Стьюдента о равенстве средних двух независимых выборок.
Генеральные совокупности не зависят друг от друга; так как выборки разного размера укажем *equal_var=False*.
Уровень статистической значимости зададим как *alpha = .05*

### Гипотеза о различии дохода между теми клиентами, которые ушли, и теми, которые остались

H0 - между средними дохода клиентов, которые ушли, и клиентов, которые остались, нет статистически значимой разницы
H1 - между средними дохода клиентов, которые ушли, и клиентов, которые остались, есть разница

```python
alpha = 0.05

result, p_value = stats.ttest_ind(data.query('churn == 0')['est_salary'], 
                                  data.query('churn == 1')['est_salary'], equal_var=False)

print('p_value: {}'.format(p_value))
if p_value < alpha:
    print('Отвергаем нулевую гипотезу: между средними дохода есть статистически значимая разница')
if p_value > alpha:
    print('Не получилось отвергнуть нулевую гипотезу, нет оснований считать доходы разными')
```

Нулевую гипотезу не отвергаем, статистически значимой разницы нет.

### Гипотеза о различии баланса между теми клиентами, которые ушли, и теми, которые остались

H0 - между средними баланса клиентов, которые ушли, и клиентов, которые остались, нет статистически значимой разницы
H1 - между средними баланса клиентов, которые ушли, и клиентов, которые остались, есть разница

```python
result, p_value = stats.ttest_ind(data.query('churn == 0')['balance'], 
                                  data.query('churn == 1')['balance'], nan_policy='omit', equal_var=False)

print('p_value: {}'.format(p_value))
if p_value < alpha:
    print('Отвергаем нулевую гипотезу: между средними баланса есть статистически значимая разница')
if p_value > alpha:
    print('Не получилось отвергнуть нулевую гипотезу, нет оснований считать балансы разными')
```

Получилось отвергнуть нулевую гипотезу, есть статистически значимая разница между балансом
